```python
# 유튜브 랭크 웹사이트: https://youtube-rank.com/board/bbs/board.php?bo_table=youtube
#title /category /subscriber /view /video 

#BeautifulSoup을 이용한 벅스뮤직 지니차트 크롤링 
from selenium import webdriver
from bs4 import BeautifulSoup
import time 

path = "/usr/local/bin/chromedriver"
driver= webdriver.Chrome(path)

url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube"
driver.get(url)

html = driver.page_source
soup= BeautifulSoup(html,"html.parser") 

```

```python
channel_data= [] 
rank=1 
channels= soup.select('tbody > tr.aos-init') 
channel= channels[0]

title = channel.select('td.subject > h1 > a')[0].text.strip()
# print(len(title))
#title /category /subscriber /view /video 

#데이터 추출 for 문
channels= soup.select('tbody > tr.aos-init') 
for channel in channels: 
    title = channel.select('td.subject > h1 > a')[0].text.strip()
    category = channel.select('td > h1 > p.category')[0].text.strip()
    subscriber= channel.select('td.subscriber_cnt')[0].text.strip() 
    view= channel.select('td.view_cnt')[0].text.strip()
    video= channel.select('td.video_cnt')[0].text.strip() 
    channel_data.append(['Youtube',title,category,subscriber,view,video])
    print(title, category,subscriber,view,video) 
    
import pandas as pd  
columns = ['서비스', 'title','category','subscriber','view','video'] 
pd_df= pd.DataFrame(channel_data, columns= columns)
pd_df
pd_df.to_excel("data/youtube.xlsx",index= False)
```

```python
##유튜브 랭킹 화면의 여러페이지 크롤링 
from selenium import webdriver
from bs4 import BeautifulSoup
import time 

path = "/usr/local/bin/chromedriver"
driver= webdriver.Chrome(path)

result = [] 
for page in range(1,11): 
    url = f"https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={page}"
    driver.get(url)
    time.sleep(10)
    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser") 
    channels= soup.select('tbody > tr.aos-init') 
    for channel in channels: 
        title = channel.select('td.subject > h1 > a')[0].text.strip()
        category = channel.select('td > h1 > p.category')[0].text.strip()
        subscriber= channel.select('td.subscriber_cnt')[0].text.strip() 
        view= channel.select('td.view_cnt')[0].text.strip()
        video= channel.select('td.video_cnt')[0].text.strip() 
        data=['Youtube',title,category,subscriber,view,video]
        result.append(data)

```

```python
df = pd.DataFrame(result)
df.columns = ['service','title','category','subscriber','view','video']
df.to_excel("data/youtube_rank.xlsx",index= False)
```

```python
df= pd.read_excel("data/youtube_rank.xlsx")
df['subscriber'].str.replace('만','0000')[0:10] 
df['replace_subscriber']= df['subscriber'].str.replace('만','0000')
df['replace_subscriber']= df['replace_subscriber'].astype('int')
df.info() #Dtype이 Object 에서 int 로 변경 
df
```

```python
##카테고리별 구독자수, 채널 수 피봇테이블 생성
pivot_df= df.pivot_table(index= 'category', values= 'replace_subscriber', aggfunc=["sum","count"])
pivot_df.head() 
```

```python
# 데이터프레임의 인덱스 초기화 
pivot_df.columns =['subscriber_sum','category_count'] 
pivot_df.head()  

pivot_df= pivot_df.reset_index() 
pivot_df.head() 

#데이터프레임 내림차순 정렬
pivot_df= pivot_df.sort_values(by='subscriber_sum',ascending=False) 
pivot_df.head() 

```

```python
pivot_df1=pivot_df.sort_values(by='category_count',ascending=False)
pivot_df1.head() 
```

```python
import matplotlib.pyplot as plt                                                            
from matplotlib import font_manager, rc

## 카테고리별 구독자수-"카테고리별 구독자 수"
plt.figure(figsize=(35,10)) 
plt.pie(pivot_df['subscriber_sum'],labels=pivot_df['category'],autopct= '%1.1f%%')
plt.show() 

## 카테고리별 채널수-"카테고리별 카테고리 채널 수"
plt.figure(figsize=(35,10))                                               ##%%로 묶고 %1.1f%%
plt.pie(pivot_df1['category_count'],labels=pivot_df1['category'],autopct= '%1.1f%%')
plt.show() 
```
