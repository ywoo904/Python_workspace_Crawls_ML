```python
#웹상의 정보수집 방법 
# 파이썬 urllib 모듈을 사용하여 정보수집 

## urllib.request 모듈은 웹사이트에 있는 데이터에 접근할 수 있는 기능을 제공 
# 인증, 리다이렉트, 쿠키처럼 인터넷을 이용한 다양한 요청과 처리를 지원함 

# 파일다운로드, urllib.request 모듈의 urlretrieve() 함수를 사용 
# 직접 다운로드 가능 

# 예시
# 라이브러리 
import urllib.request

# 접속경로 및 저장이름 
url = "https://uta.pw/shodou/img/28/214.png" 
savename= "test_download.png"

# 다운로드 
urllib.request.urlretrieve(url,"/Users/wyattwoo/Desktop/Python_Workspace/Python_Workspace/day001/download/"+savename) 

# 네이버 로고를 다운로드 해보자 
url1 = "http://s.pstatic.net/static/www/mobile/edit/20220503/mobile_20183742258.gif"
savename1= "naver_logo.gif"
urllib.request.urlretrieve(url1,"/Users/wyattwoo/Desktop/Python_Workspace/Python_Workspace/day001/download/"+savename1) 





```

```python
## API 접속 

# 모듈  
import urllib.request 

# 데이터 읽기  
url = "https://api.aoikujira.com/ip/ini" 
res = urllib.request.urlopen(url) 
data = res.read() # 저장된 데이터를 read () 내장함수로 읽는다. 
print(data)
# \n이 붙은 것으로 보아 binary 값으로 저장, 문자열로 디코딩을 해야겠다. 
# \nAPI_URI=http://api.aoikujira.com/ip/get.php\nREMOTE_ADDR=14.39.239.87\nREMOTE_HOST=14.39.239.87\nREMOTE_PORT=57836\nHTTP_HOST=api.aoikujira.com\nHTTP_USER_AGENT=Python-urllib/3.10\nHTTP_ACCEPT_LANGUAGE=\nHTTP_ACCEPT_CHARSET=\nSE

text = data.decode("utf-8")
print(text)
```

```python
# 기상청 RSS URL, 파일형식: xml  
url= "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" 

# 매개변수: stnId - 기상정보를 알고 싶은 지역을 지정 
# ex) https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnid=109 서울.인천.경기도 
## 전국-108 / 서울경기-109 / 강원도-105 / 충청북도-131 / 충청남도-133 / 전라북도-146 / 전라남도-156 / 경상북도-143 / 경상남도-159 / 제주도-184

import urllib.request 
import urllib.parse 
API = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" 

# 매개변수를 url 인코딩 
values = { 
    "stnId": "184"  # 키값: 변수명 / 벨류값: 변수값  
}

params = urllib.parse.urlencode(values)

url = API + "?" + params

# 다운로드 
data= urllib.request.urlopen(url).read() 
text= data.decode("utf-8") 
print(text)

```

```python
## 입력값(지역번호)를 받아서 해당 지역의 날씨정보를 받아오세요 
# sys module...확인하기 
local = int(input("지역번호를 입력하세요")) 

import urllib.request 
import urllib.parse 
API = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" 

# 매개변수를 url 인코딩 , **매개변수가 여러개일 경우에는 접속을 여러번 시행하는 것으로 반복문을 실행해야한다 
values = { 
    "stnId": local,  # 키값: 변수명 / 벨류값: 변수값  
}

params = urllib.parse.urlencode(values) #매개변수를 url 인코딩 

url = API + "?" + params 
print(url) # https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=184 

# 다운로드 
data= urllib.request.urlopen(url).read() 
text= data.decode("utf-8") 
print(text) 
```
