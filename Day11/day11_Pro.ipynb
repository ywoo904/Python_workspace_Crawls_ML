### 파일의 확장자를 파일명.ipynb

#### 관련용어

 1. 클롤링
 
 크롤링(Crawling)이란, 웹을 탐색하는 컴퓨터 프로그램(크롤러)을 이용하여 여러 인터넷 사이트의 웹페이지 자료를 수집하여 분류하는 과정을 말함.
 
 2. 스크래핑
 
 스크래핑(Scraping)이란, 웹 사이트의 내용을 긁어다 원하는 형태로 가공하는 기술을 의미한다. 즉, 웹 사이트의 데이터를 수집하는 모든 작업을 의미함. 
 
 3. 파싱
 
 파싱(parsing)이란, 어떤 페이지(문서, html 등)에서 내가 원하는 데이터를 특정 패턴이나 순서로 추출하여 정보를 가공하는 것을 말함. 예로 html소스를 문자열로 수집하여 html태그로 인식하도록 정보를 가종하여 html단위별 분석이 가능하게 구성할 수 있음. 

```python
import urllib.request   # 웹에 통신 데이터를 요청하는 모듈.
from bs4 import BeautifulSoup

## html source 가져오기... (www.naver.com)
url = 'http://www.naver.com'
url2 = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100' 

resp = urllib.request.urlopen(url2)
data = resp.read()

## html 파싱
html = data.decode("euc-kr")  # byte -> 문서
type(html) # str
soup = BeautifulSoup(html,"html.parser")  # html 문서를 html소스로 파싱

## 태그 내용 가져오기
# 1. h1태그 가져오기(find()메서드 사용과 직접 참조)
h1 = soup.find('h1')
print(h1)
print('-'*50)
print(h1.a)
print('-'*50)
print(h1.a.string)

# 2. find_all() 
h2s = soup.find_all('h2')

print("-"*50)
print("h2태그 문자열들을 출력===>")
for h2 in h2s:
    print(h2.string)
```

```python
## 파일을 이용하여 html01.html을 읽어보세요.... 
import urllib.request
from bs4 import BeautifulSoup

import os

os.getcwd()
## html01.html 파일읽기
rfile = open('data/html01.html','r',encoding='utf8')
html01 = rfile.read()

# html 파싱
source = BeautifulSoup(html01,'html.parser')

## h1 접근
h1 = source.html.body.h1    # DOM
print(h1.string)

## h2 접근 : find()
h2 = source.find('h2')
print('h2 :',h2.string)

## body접근 children
body = source.html.body
print('body태그 하위 태그들')
for i in body.children:
    print(i)

## li태그들을 찾아서 출력하고, 내용을 출력하세요. 
litags = source.find_all('li')
litags
idx = 0
for li in litags:
    print(f"litags[{idx}] : {li}")
    idx += 1
```

```python
# html03.html에서 선택자를 이용한 정보수집
from bs4 import BeautifulSoup

# 로컬 파일 읽기
rfile3 = open('data/html03.html','r',encoding='utf-8')
src3 = rfile3.read()

# html 파싱
html3 = BeautifulSoup(src3,'html.parser')

# 선택자를 이용한 내용 가져오기(select, select_one)
# 1. id = 'tab' 선택자 정보 가져오기
print(">>> table 선택자 <<<")
table = html3.select_one('#tab')  # table에 있는 id값 'tab'으로 접근
# print(table)  #table 태그 전체 출력 

# 2. id선택자와 계층
print(">>> 선택자 & 계층 <<<")
ths = html3.select('#tab > tr > th')
print(ths)    # list형태로 반환

# 3. class 선택자
print(">>> 클래스 선택자 <<<<")
trs = html3.select('#tab > .odd')
print(trs)

print(">>> 속성값을 이용한 선택 <<<")
trs2 = html3.select('tr[class="odd"]')
print(trs2)

### tr > td에 있는 문자열을 출력하는 코드를 작성하세요.
tds = html3.select('#tab > tr > td')
print(">>> tr > td 출력 <<<")
for td in tds:
    print(td.string)
```
