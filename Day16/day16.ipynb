```python
import pandas as pd

jumsu=  { 
'홍길동' : [50,60,50], 
'이순신' : [60,60,60],
'강감찬' : [100,90,90], 
'유관순' : [80,80,90],
'강환석' : [95,95,95]
}  

#4개의 Series로 만들어주세요 
#이름(name_s), 각 리스트 0번은 국어(kor_S), 리스트 1번은 영어(eng_S),
#리스트 2번은 수학(math_s)

#이름 -> 시리즈 이름
# 이름 시리즈
na = [key for key in jumsu]
name_s = pd.Series(na)
# 국어 시리즈
kor = [jumsu[key][0] for key in na]
kor_s = pd.Series(kor)
# 영어 시리즈
eng_s = pd.Series([jumsu[key][1] for key in na])
# 수학 시리즈
math_s = pd.Series([jumsu[key][2] for key in na])

print(name_s) 
print(kor_s)
print(eng_s)
print(math_s)

# DataFrame 만들자 ... 
score = pd.DataFrame({'이름': name_s,'국어': kor_s,'영어':eng_s,'수학':math_s })
print(score)

##국어 영어 수학 합계 시리즈를 생성하세요 
total= kor_s+ eng_s+ math_s
print("===총합계===")
print(total)

## 평균값을 구하기  
avg= round(total/3,2) 
print("===평균===")
print(avg)

score['합계']= total
score['평균']= avg
print(score)
```
